```python
import pyedflib
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
full_edf_file_path = os.path.join('data/test_dir/raw_data/edfs/abc-baseline-900001.edf')
f = pyedflib.EdfReader(full_edf_file_path)
  
    
channel_list = f.getSignalLabels()
# ecg_channel_name = [channel for channel in channel_list if self.ecg_channel_regex.match(channel)][0]
ecg_channel_name_one = 'ECG1' #'ECG2'
ecg_channel_one = channel_list.index(ecg_channel_name_one)
ecg_signal_one = f.readSignal(ecg_channel_one )
ecg_freq_one = f.getSampleFrequency(ecg_channel_one)

ecg_channel_name_two = 'ECG2' #'ECG2'
ecg_channel_two = channel_list.index(ecg_channel_name_two)
ecg_signal_two = f.readSignal(ecg_channel_two )
ecg_freq_two = f.getSampleFrequency(ecg_channel_two)
f._close()
```
